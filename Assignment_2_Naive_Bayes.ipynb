```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


# download the dataset
path = kagglehub.dataset_download("chitwanmanchanda/fraudulent-transactions-data")
print(path)

df = pd.read_csv(path + "/Fraud.csv")
df.head()
```

```python
# size of dataset
print(df.shape)

# info and data types
print(df.info())

# Checking for missing values
print(df.isnull().sum())
```

```python
# some basic stats
print(df.describe())
```

```python
# how many fraud vs not fraud
df['isFraud'].value_counts()
```

```python
# percentage of fraud
fraud_percentage = df['isFraud'].mean()*100
print(fraud_percentage)
```

```python
# checking for missing values
print(df.isnull().sum())
```

```python
# dropping account name columns
df = df.drop(columns=['nameOrig', 'nameDest'],errors='ignore')

print(df.columns)
```

```python
# turning transaction type into dummy columns
X = df.drop(columns=['isFraud'])
y = df['isFraud']

X = pd.get_dummies(X, columns=['type'], drop_first=False)

# Checking columns
print(X.columns)
print(y.name)

pd.concat([X, y], axis=1).head(10)
```

```python
from sklearn.model_selection import train_test_split

# splitting data 70/30
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

# checking
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

# making the model
model = GaussianNB()

# training the model
model.fit(X_train, y_train)
```

```python
# Probability of each class in training data
print(model.class_prior_)

# How many features the model is using
print(model.n_features_in_)
```

```python
# making predictions
y_pred = model.predict(X_test)

# checking first 10
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
